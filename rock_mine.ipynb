libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and data Processing

```python
#loadig dataset into pandas dataframe
sonar_data = pd.read_csv('/content/Copy of sonar data.csv',header = None)
```

```python
sonar_data.head()
```

```python
#number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
#M--> mine R-->rock
sonar_data.groupby(60).mean()
```

seperating data and lables

```python
x = sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
print(x)
print(y)
```

train and test data


```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
print(y_train)
```

model training ==> logestic algorithm

```python
model = LogisticRegression()
```

```python
#training the logestic regression model with trainng data
model1=LogisticRegression().fit(x_train , y_train)
```

model evalution

```python
#accuracy of traing data
x_train_prediction = model1.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print(training_data_accuracy )
```

```python
#accuracy of test data
x_test_predection = model1.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predection,y_test)
```

```python
print(test_data_accuracy )
```
