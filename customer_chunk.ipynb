```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
import pickle
```

```python
dataset=pd.read_csv('/content/Churn_Modelling.csv')
dataset.head()
```

```python
x=dataset.iloc[:,3:-1]
y=dataset.iloc[:,-1].values
x.head()
```

```python
y
```

```python
print(x['Geography'].value_counts())
print(x['Gender'].value_counts())
print(x['NumOfProducts'].value_counts())
print(x['HasCrCard'].value_counts())
print(x['IsActiveMember'].value_counts())
```

```python
x.describe()
```

```python
#isnull shows all value in true or false sum is for calcluate the count of true
x.isnull().sum()
```

```python
unique, counts=np.unique(y, return_counts=True)

print(np.asarray((unique, counts)))
```

```python
#values of number into 0 and 1 using one hand encoding
x=pd.get_dummies(x,drop_first=True)
```

```python
#training the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1,stratify=y,random_state=1)
print(x_train)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
print(x_train)
```

```python
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
pickle.dump(rf,open('customer_chunk_pred.pkl','wb'))
```
